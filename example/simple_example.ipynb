```python
import sys
import os
import torch
import gymnasium as gym #read this https://gymnasium.farama.org/introduction/basic_usage/
import ale_py
#cwd = os.getcwd()
#file_path = os.path.abspath(os.path.join(os.path.dirname(cwd), '.', 'src', 'components'))
#os.chdir(file_path)
import /src/components/ReplayMemory as rm
#os.chdir(cwd)

gym.register_envs(ale_py) #register ale with gymnasium api

env = gym.make('ALE/Breakout-v5') #create the enviroment
observation, info = env.reset() #get initial info of enviroment

episode_over = False #flag to see if we lost
total_reward = 0 #reward is a measure of how good we are doing
replay_memory = rm.ReplayMemory(1000)
count = 0    

while not episode_over: #main loop
    
    action = env.action_space.sample() #randomly samples available actions this is eventually the 'policy'

    obs, rwrd, trmt, trnc, info = env.step(action) #step in enviroment get next observation, recieve reward and see if the episode ended or truncated
    replay_memory.push(observation, action, obs, rwrd)
    observation = obs
    
    total_reward += rwrd #log reward gained for action
    episode_over = trmt or trnc #break if we lost
    count += 1
env.close()
```
